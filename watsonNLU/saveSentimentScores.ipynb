```python
import csv
import json
from watson_developer_cloud import NaturalLanguageUnderstandingV1
from watson_developer_cloud.natural_language_understanding_v1 \
  import Features, EntitiesOptions, KeywordsOptions, SentimentOptions
    
    
def sentimentAnalysis(filename):
    # Analyzes every entry's (title + description) with Bluemix Natural Language Understadning 
    # and appends sentiment value to the data entry
    # Saves the new file in: filename_withScores.csv
    
    nlu = NaturalLanguageUnderstandingV1(
      username='40de6523-64fc-40b5-a87c-4e1d674f02e9',
      password='nms1o55i61mL',
      version='2018-03-16')
    i = 0
    with open(filename) as f:
        reader = csv.reader(f)
        header = next(reader)
        with open(filename.split(".csv")[0] + "_BluemixScore.csv", "w") as file_out:
            writer = csv.writer(file_out)
            header = ["Title", "Author", "Publish Date", "Publisher", "Description", "url", "TitleScore"]
            writer.writerow(header)
            for row in reader:
                print(row)
                article = row[0] + " " + row[4]
                if "bp" in article.lower():
                    SentOpt = SentimentOptions(
                    targets = ["bp"])
                else:
                    SentOpt = SentimentOptions(
                    document=True)
                response = nlu.analyze(
                    text = article,
                    #html = article_html,
                    features=Features(
                        entities=EntitiesOptions(
                          emotion=True,
                          sentiment=True,
                          limit=2),

                        sentiment=SentOpt
                    ))

                print(article)
                #print(response)
                if 'targets' in response['sentiment']:
                    print("The sentiment is: " + str(response["sentiment"]))
                    print(response["sentiment"]['targets'][0]['score'])
                    #row.append(response["sentiment"]['targets'][0]['score'])
                    row[6] = response["sentiment"]['targets'][0]['score']
                else:
                    print("There were no targets, but the document score is: " + str(response['sentiment']))
                    print(response['sentiment']['document']['score'])
                    #row.append(response['sentiment']['document']['score'])
                    row[6] = response['sentiment']['document']['score']
                print("-----------------------------")
                writer.writerow(row)
                #print(article)
            
```
