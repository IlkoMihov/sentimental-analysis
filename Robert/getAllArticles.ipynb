```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import requests

# Open connection, grabs the page
def getSoup(url):
    uClient = uReq(url)
    page_html = uClient.read()
    #print(page_html)
    uClient.close()
    return soup(page_html, "html.parser")  
```
